```python
import csv
import requests

# Fetch the JSON data from the API
url = 'https://www.sofascore.com/api/v1/event/12436513/shotmap'
#headers = {"User-Agent": "YourUserAgentHere"}  # Replace with a valid User-Agent if required
response = requests.get(url, headers=headers)
shots = response.json()

# Specify the file name for the CSV
csv_file = 'shots.csv'

# Open the file and write the data
with open(csv_file, mode='w', newline='', encoding='utf-8') as file:
    writer = csv.writer(file)
    
    # Write the header row (update field names based on JSON structure)
    writer.writerow(['playerName', 'x', 'y', 'type', 'onTarget'])  # Example fields
    
    # Write the data rows
    for shot in shots.get('shots', []):  # Update the key based on actual JSON structure
        writer.writerow([
            shot.get('playerName'),  # Replace with the correct field
            shot.get('x'),
            shot.get('y'),
            shot.get('type'),
            shot.get('onTarget')
        ])
        
print(f"Data saved to {csv_file}")
```
